```python
from kinematics import *

leg_list = LegList()
leg1 = Leg()
```

```python
import numpy as np

d2r = np.deg2rad
```

```python
element = 100

hip_pitch_range = np.linspace(d2r(90),d2r(150), element)
knee_range = np.linspace(d2r(70), d2r(120), element)
```

```python
px = [0]
py = [0]

for index1 in range(len(hip_pitch_range)):
    for index2 in range(len(knee_range)):
        motor1_angle = hip_pitch_range[index1]
        motor2_angle = knee_range[index2]

        leg1.move_motor(motor1_angle, 0, False)
        leg1.move_motor(motor2_angle, 1, False)
        leg1.update()
        result_position = leg1.segments[2].abs_end_coords()

        px.append(result_position[0])
        py.append(result_position[1])
```

```python
%matplotlib ipympl
import numpy as np
import matplotlib.pyplot as plt 

plt.close()

fig = plt.figure()
 
# syntax for 3-D projection
ax = fig.add_subplot(111)

xs = np.array(px)
ys = np.array(py)
 
# plotting
ax.plot(px, py, 'o', markersize=2, color='orange')
ax.set_title('')

plt.grid()
plt.show()
```
